```python
import numpy as np
import pandas as pd
import re
import string
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    confusion_matrix
)

# =========================================================
# LOAD DATA
# =========================================================

train_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/train.csv'
)

test_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/test.csv'
)

sample_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/sample_submission.csv'
)

# =========================================================
# CLEAN COLUMN NAMES
# =========================================================

train_data.columns = train_data.columns.str.strip()
test_data.columns = test_data.columns.str.strip()

# =========================================================
# TEXT CLEANING FUNCTION
# =========================================================

def clean_text(text):

    text = str(text).lower()

    # remove urls
    text = re.sub(r'http\S+|www\S+', '', text)

    # remove html tags
    text = re.sub(r'<.*?>', '', text)

    # remove numbers
    text = re.sub(r'\d+', '', text)

    # remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# =========================================================
# APPLY CLEANING
# =========================================================

train_data['review'] = train_data['review'].apply(clean_text)
test_data['review'] = test_data['review'].apply(clean_text)

# =========================================================
# FEATURES AND TARGET
# =========================================================

X = train_data['review']
y = train_data['rating']

X_test_final = test_data['review']

# =========================================================
# TRAIN TEST SPLIT
# =========================================================

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("Train Shape :", X_train.shape)
print("Validation Shape :", X_valid.shape)

# =========================================================
# TFIDF + LOGISTIC REGRESSION PIPELINE
# =========================================================

model = Pipeline([
    
    (
        'tfidf',
        TfidfVectorizer(
            
            stop_words='english',
            
            max_features=50000,

            ngram_range=(1, 2),

            min_df=3,

            max_df=0.90,

            sublinear_tf=True,

            strip_accents='unicode',

            token_pattern=r'[a-zA-Z]+'
        )
    ),

    (
        'classifier',
        LogisticRegression(
            max_iter=3000,
            class_weight='balanced',
            C=2.0,
            solver='liblinear',
            random_state=42
        )
    )
])

# =========================================================
# TRAIN MODEL
# =========================================================

model.fit(X_train, y_train)

# =========================================================
# VALIDATION PREDICTION
# =========================================================

y_pred = model.predict(X_valid)

# =========================================================
# EVALUATION
# =========================================================

print("\nAccuracy Score :")
print(accuracy_score(y_valid, y_pred))

print("\nClassification Report :")
print(classification_report(y_valid, y_pred))

print("\nConfusion Matrix :")
print(confusion_matrix(y_valid, y_pred))

# =========================================================
# TEST PREDICTION
# =========================================================

test_prediction = model.predict(X_test_final)

# =========================================================
# SUBMISSION FILE
# =========================================================

submission = pd.DataFrame({
    'id': sample_data['id'],
    'rating': test_prediction
})

submission.to_csv('submission.csv', index=False)

print("\nsubmission.csv file created successfully")
```

```python
# =========================================================
# IMPORT LIBRARIES
# =========================================================

import numpy as np
import pandas as pd
import torch
import re
import string
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    f1_score
)

from datasets import Dataset

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding
)

# =========================================================
# LOAD DATA
# =========================================================

train_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/train.csv'
)

test_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/test.csv'
)

sample_data = pd.read_csv(
    '/kaggle/input/competitions/mlp-term-2-2025-kaggle-assignment-3/sample_submission.csv'
)

# =========================================================
# CLEAN COLUMN NAMES
# =========================================================

train_data.columns = train_data.columns.str.strip()
test_data.columns = test_data.columns.str.strip()

# =========================================================
# TEXT CLEANING FUNCTION
# =========================================================

def clean_text(text):

    text = str(text).lower()

    text = re.sub(r'http\S+|www\S+', '', text)

    text = re.sub(r'<.*?>', '', text)

    text = re.sub(r'\d+', '', text)

    text = text.translate(str.maketrans('', '', string.punctuation))

    text = re.sub(r'\s+', ' ', text).strip()

    return text

# =========================================================
# APPLY CLEANING
# =========================================================

train_data['review'] = train_data['review'].apply(clean_text)
test_data['review'] = test_data['review'].apply(clean_text)

# =========================================================
# LABEL ENCODING
# BERT expects labels from 0
# =========================================================

train_data['label'] = train_data['rating'] - 1

# =========================================================
# TRAIN VALID SPLIT
# =========================================================

train_df, valid_df = train_test_split(
    train_data,
    test_size=0.20,
    stratify=train_data['label'],
    random_state=42
)

print("Train Shape :", train_df.shape)
print("Valid Shape :", valid_df.shape)

# =========================================================
# CONVERT TO HF DATASET
# =========================================================

train_dataset = Dataset.from_pandas(
    train_df[['review', 'label']]
)

valid_dataset = Dataset.from_pandas(
    valid_df[['review', 'label']]
)

test_dataset = Dataset.from_pandas(
    test_data[['review']]
)

# =========================================================
# LOAD TOKENIZER
# =========================================================

MODEL_NAME = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# =========================================================
# TOKENIZATION FUNCTION
# =========================================================

def tokenize_function(example):

    return tokenizer(
        example["review"],
        truncation=True,
        padding='max_length',
        max_length=256
    )

# =========================================================
# TOKENIZE DATA
# =========================================================

train_dataset = train_dataset.map(tokenize_function, batched=True)
valid_dataset = valid_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)

# =========================================================
# DATA COLLATOR
# =========================================================

data_collator = DataCollatorWithPadding(
    tokenizer=tokenizer
)

# =========================================================
# LOAD MODEL
# =========================================================

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=5
)

# =========================================================
# METRICS FUNCTION
# =========================================================

def compute_metrics(eval_pred):

    logits, labels = eval_pred

    predictions = np.argmax(logits, axis=-1)

    accuracy = accuracy_score(labels, predictions)

    macro_f1 = f1_score(
        labels,
        predictions,
        average='macro'
    )

    return {
        "accuracy": accuracy,
        "macro_f1": macro_f1
    }

# =========================================================
# TRAINING ARGUMENTS
# =========================================================

training_args = TrainingArguments(

    output_dir="./results",

    eval_strategy="epoch",

    save_strategy="epoch",

    learning_rate=2e-5,

    per_device_train_batch_size=16,

    per_device_eval_batch_size=16,

    num_train_epochs=3,

    weight_decay=0.01,

    load_best_model_at_end=True,

    metric_for_best_model="macro_f1",

    logging_steps=100,

    fp16=torch.cuda.is_available(),

    report_to="none"
)

# =========================================================
# TRAINER
# =========================================================

trainer = Trainer(

    model=model,

    args=training_args,

    train_dataset=train_dataset,

    eval_dataset=valid_dataset,

    data_collator=data_collator,

    compute_metrics=compute_metrics
)

# =========================================================
# TRAIN MODEL
# =========================================================

trainer.train()

# =========================================================
# VALIDATION PREDICTION
# =========================================================

predictions = trainer.predict(valid_dataset)

y_pred = np.argmax(predictions.predictions, axis=1)

y_true = valid_df['label'].values

# =========================================================
# EVALUATION
# =========================================================

print("\nAccuracy Score :")
print(accuracy_score(y_true, y_pred))

print("\nMacro F1 Score :")
print(f1_score(y_true, y_pred, average='macro'))

print("\nClassification Report :")
print(classification_report(y_true, y_pred))

print("\nConfusion Matrix :")
print(confusion_matrix(y_true, y_pred))

# =========================================================
# TEST PREDICTION
# =========================================================

test_predictions = trainer.predict(test_dataset)

test_pred_labels = np.argmax(
    test_predictions.predictions,
    axis=1
)

# Convert back to original labels
test_pred_labels = test_pred_labels + 1

# =========================================================
# SUBMISSION FILE
# =========================================================

submission = pd.DataFrame({
    'id': sample_data['id'],
    'rating': test_pred_labels
})

submission.to_csv('submission.csv', index=False)

print("\nsubmission.csv created successfully")
```
